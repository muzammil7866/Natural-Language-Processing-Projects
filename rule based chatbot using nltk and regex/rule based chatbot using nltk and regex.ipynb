```python
import nltk
import random
import string

from nltk.chat.util import Chat, reflections

# Download required NLTK data
nltk.download('punkt')

# Define some basic patterns and responses
pairs = [
    [
        r"hi|hello|hey",
        ["Hello!", "Hi there!", "Hey! How can I help you?"]
    ],
    [
        r"what is your name ?",
        ["I am a chatbot created for NLP demo!", "You can call me NLPBot."]
    ],
    [
        r"how are you ?",
        ["I'm doing good, thank you!", "Feeling great, how about you?"]
    ],
    [
        r"(.*) your name ?",
        ["My name is NLPBot.", "I'm called NLPBot."]
    ],
    [
        r"what can you do ?",
        ["I can chat with you using basic NLP.", "Right now, I'm just a demo bot."]
    ],
    [
        r"quit",
        ["Bye! Take care.", "Goodbye!"]
    ],
    [
        r"(.*)",
        ["Sorry, I didn't understand that.", "Can you rephrase that?"]
    ]
]

# Create chatbot
chatbot = Chat(pairs, reflections)

# Start chat
print("Hi! I'm NLPBot. Type 'quit' to exit.")
chatbot.converse()
```
